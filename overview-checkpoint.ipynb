```python
# Returns tuple with (true if an element of keyword is in the search string, number of elements that match)
def contains(keywords, string):
    keywords = set(keywords)
    string = set(string.lower().split(' '))

    in_both = keywords.intersection(string)

    return (len(in_both) > 0, len(in_both))
```

```python
# first filter out websites by blacklist (porn sites etc)
english_sites['blacklisted'] = english_sites['meta_desc'].apply(lambda text: 
                                                               contains(blacklist.split(' '),  str(text))[0])
# english_sites['blacklisted']
remaining_sites = english_sites[english_sites['blacklisted'] == False]
print(len(remaining_sites))
```

```python
# try filtering just on meta_desc, bc most do not have an all_text field
# use boolean (first element of tuple) get lots of "children sites"
# import utils.filters as ufi

english_sites['kids'] = remaining_sites['meta_desc'].apply(lambda text:
                                            contains(keywords.split(' '),  str(text))[0])
kid_sites = english_sites[english_sites['kids'] == True]
kid_sites
```

```python
# use count greater than 2 get lots less children sites and they're more accurate 
# same as above, but for comparison of 1 vs 2 matches
# replace english_sites['meta_desc'] with remaining_sites['meta_desc']

english_sites['count'] = remaining_sites['meta_desc'].apply(lambda text:
                                            contains(keywords.split(' '),  str(text))[1])
kid_sites = english_sites[english_sites['count'] >= 2]
kid_sites
```
